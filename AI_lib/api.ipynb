```python
import xml.etree.ElementTree as ET
import requests

url = "https://apis.data.go.kr/B552657/HsptlAsembySearchService/getBabyLcinfoInqire?serviceKey=UhIXqz3BxyzUBIOobK%2ByC%2BlbhlKx4lOWT5glOai8uSuRyW%2Fea2D6OBF7zp5WnyjTYcXyrzk%2BPXV2KtNrChOzOQ%3D%3D&pageNo=1&numOfRows=10&WGS84_LON=127.0851566&WGS84_LAT=37.48813256"


response = requests.get(url)
contents = response.text

# XML 데이터를 파싱하여 Element 객체 생성
root = ET.fromstring(contents)

# 보기 좋게 출력하기 위해 들여쓰기를 추가한 XML 데이터 생성
def indent(elem, level=0):
    indent_str = '\n' + level * '\t'
    if len(elem):
        if not elem.text or not elem.text.strip():
            elem.text = indent_str + '\t'
        if not elem.tail or not elem.tail.strip():
            elem.tail = indent_str
        for elem in elem:
            indent(elem, level + 1)
        if not elem.tail or not elem.tail.strip():
            elem.tail = indent_str
    else:
        if level and (not elem.tail or not elem.tail.strip()):
            elem.tail = indent_str

indent(root)

# 보기 좋게 정리된 XML 데이터 출력
print(ET.tostring(root, encoding='unicode'))
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
import requests

url = "https://apis.data.go.kr/B552657/HsptlAsembySearchService/getBabyLcinfoInqire?serviceKey=UhIXqz3BxyzUBIOobK%2ByC%2BlbhlKx4lOWT5glOai8uSuRyW%2Fea2D6OBF7zp5WnyjTYcXyrzk%2BPXV2KtNrChOzOQ%3D%3D&pageNo=1&numOfRows=10&WGS84_LON=127.0851566&WGS84_LAT=37.48813256"


response = requests.get(url)
contents = response.text

# XML 데이터를 파싱하여 Element 객체 생성
root = ET.fromstring(contents)

# 필요한 정보 추출하여 리스트에 저장
data = []
for item in root.iter('item'):
    distance = float(item.find('distance').text)
    address = item.find('dutyAddr').text
    name = item.find('dutyName').text
    tell = item.find('dutyTel1').text
    startTime = item.find('startTime').text[:2] + ":" + item.find('startTime').text[2:]
    endTime = item.find('endTime').text[:2] + ":" + item.find('endTime').text[2:]
    latitude = float(item.find('latitude').text)
    longitude = float(item.find('longitude').text)
    data.append({'거리': distance, '주소': address, '이름': name, '전화번호': tell, 'Open': startTime, 'Close': endTime, '위도': latitude, '경도': longitude})

# 추출된 정보를 데이터프레임으로 변환
df = pd.DataFrame(data)

# 데이터프레임 출력
df
```
